```python
# ============================================================
#  🧠 TREINO — ResNet-50 | Retinopatia Diabética
#  📦 Dataset: augmented_resized_V2 (143.687 imagens)
#  🔄 Retomada automática de checkpoint
#  👉 Cole tudo em UMA célula do Colab e rode
# ============================================================
# ⚡ ANTES DE RODAR:
#    Ambiente de execução → Alterar tipo de execução → GPU (T4)
# ============================================================

# ── 1. Instalar dependências ─────────────────────────────────
import subprocess
subprocess.run(['pip', 'install', '-q', 'timm', 'albumentations',
                'opencv-python', 'scikit-learn', 'tqdm', 'seaborn'])

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# ── 2. Imports ───────────────────────────────────────────────
import os, gc, time, warnings, zipfile
import torch
import torch.nn as nn
import torchvision.models as models
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import seaborn as sns

from torch.utils.data           import Dataset, DataLoader
from torch.cuda.amp             import GradScaler, autocast
from sklearn.model_selection    import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics            import (classification_report,
                                        confusion_matrix,
                                        roc_auc_score,
                                        accuracy_score,
                                        cohen_kappa_score)
from sklearn.preprocessing      import label_binarize
from tqdm.auto                  import tqdm
import torch.optim              as optim
import albumentations           as A
from albumentations.pytorch     import ToTensorV2

warnings.filterwarnings('ignore')
plt.rcParams.update({'figure.dpi': 110, 'font.size': 10})

# ── 3. Device ────────────────────────────────────────────────
device  = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
USE_AMP = (device.type == 'cuda')
print(f'Dispositivo : {device}')
if device.type == 'cuda':
    print(f'GPU         : {torch.cuda.get_device_name(0)}')
    print(f'VRAM        : {torch.cuda.get_device_properties(0).total_memory/1e9:.1f} GB')

# ── 4. Configurações ─────────────────────────────────────────
ZIP_PATH        = '/content/drive/MyDrive/RETINOPATIA/DATAG/augmented_resized_V2.zip'
EXTRACT_DIR     = '/content/dataset_resnet'
CHECKPOINT_DIR  = '/content/drive/MyDrive/RETINOPATIA/checkpoints_resnet50'
CHECKPOINT_PATH = os.path.join(CHECKPOINT_DIR, 'resnet50_checkpoint.pth')
FINAL_MODEL     = '/content/drive/MyDrive/RETINOPATIA/resnet50_retinopatia_final.pth'
OUTPUT_DIR      = '/content/drive/MyDrive/RETINOPATIA/resnet50_resultados'

os.makedirs(CHECKPOINT_DIR, exist_ok=True)
os.makedirs(OUTPUT_DIR,     exist_ok=True)

IMG_SIZE             = 224
BATCH_SIZE           = 32
NUM_WORKERS          = 2
EPOCHS               = 15
LR                   = 3e-4
FREEZE_EPOCHS        = 3
UNFREEZE_LR          = 5e-5
SAVE_EVERY_N_BATCHES = 500
EARLY_STOP_PATIENCE  = 5

N_CLASSES   = 5
CLASS_NAMES = ['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative']

print(f'\nConfig ResNet-50:')
print(f'  IMG={IMG_SIZE} | BATCH={BATCH_SIZE} | EPOCHS={EPOCHS}')
print(f'  LR={LR} | FREEZE={FREEZE_EPOCHS} epochs | AMP={USE_AMP}')

# ── 5. Extrair ZIP (só se necessário) ────────────────────────
if not os.path.exists(EXTRACT_DIR) or len(os.listdir(EXTRACT_DIR)) == 0:
    print(f'\n📦 Extraindo {ZIP_PATH}...')
    print('   Aguarde — 3,46 GB...')
    t0 = time.time()
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_DIR)
    print(f'   ✅ Extraído em {(time.time()-t0)/60:.1f} min')
else:
    print(f'\n✅ Dataset já extraído em {EXTRACT_DIR}')

# Inspecionar estrutura
print('\n📂 Estrutura encontrada:')
for root, dirs, files in os.walk(EXTRACT_DIR):
    depth = root.replace(EXTRACT_DIR, '').count(os.sep)
    if depth > 2:
        break
    indent  = '  ' * depth
    n_files = len([f for f in files if f.lower().endswith(('.jpg','.jpeg','.png'))])
    label   = f'  [{n_files} imagens]' if n_files > 0 else ''
    print(f'{indent}{os.path.basename(root)}/{label}')

# ── 6. Detectar raiz e montar DataFrame ──────────────────────
def find_dataset_root(base):
    for root, dirs, files in os.walk(base):
        imgs = [f for f in files if f.lower().endswith(('.jpg','.jpeg','.png'))]
        if len(imgs) > 10:
            return os.path.dirname(root)
        for d in dirs:
            sub      = os.path.join(root, d)
            sub_imgs = [f for f in os.listdir(sub)
                        if f.lower().endswith(('.jpg','.jpeg','.png'))]
            if len(sub_imgs) > 10:
                return root
    return base

DATASET_ROOT = find_dataset_root(EXTRACT_DIR)
print(f'\n📁 Raiz detectada: {DATASET_ROOT}')

subfolders = sorted([
    d for d in os.listdir(DATASET_ROOT)
    if os.path.isdir(os.path.join(DATASET_ROOT, d))
])
print(f'Pastas: {subfolders}')

NAME_MAP = {
    'No_DR':0, 'No_dr':0, 'no_dr':0, '0':0, 'normal':0, 'healthy':0,
    'Mild':1,  'mild':1,  '1':1,
    'Moderate':2, 'moderate':2, '2':2,
    'Severe':3,   'severe':3,   '3':3,
    'Proliferate_DR':4, 'Proliferative':4, 'proliferative':4,
    '4':4, 'proliferate_dr':4
}

LABEL_MAP = {f: NAME_MAP[f] for f in subfolders if f in NAME_MAP}
if not LABEL_MAP:
    LABEL_MAP = {f: i for i, f in enumerate(subfolders[:5])}

print(f'Mapeamento: {LABEL_MAP}')

data = []
for folder, lbl in LABEL_MAP.items():
    folder_path = os.path.join(DATASET_ROOT, folder)
    if not os.path.exists(folder_path):
        print(f'⚠️  Não encontrada: {folder_path}')
        continue
    for img in os.listdir(folder_path):
        if img.lower().endswith(('.jpg','.jpeg','.png')):
            data.append([os.path.join(folder_path, img), lbl])

df = pd.DataFrame(data, columns=['image','label'])
print(f'\nTotal de imagens: {len(df)}')
for i, name in enumerate(CLASS_NAMES):
    n   = (df['label'] == i).sum()
    pct = 100 * n / len(df)
    print(f'  [{i}] {name:<18} {n:>7} imagens  ({pct:.1f}%)')

assert len(df) > 0, '❌ Nenhuma imagem! Verifique ZIP_PATH.'

# ── 7. Split e class weights ─────────────────────────────────
train_df, val_df = train_test_split(
    df, test_size=0.15, stratify=df['label'], random_state=42
)
print(f'\nTreino   : {len(train_df)}')
print(f'Validação: {len(val_df)}')

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.arange(N_CLASSES),
    y=train_df['label'].values
)
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float32).to(device)

print('\nPesos de classe:')
counts = train_df['label'].value_counts().sort_index()
for i, (name, w) in enumerate(zip(CLASS_NAMES, class_weights)):
    print(f'  [{i}] {name:<18} n={counts.get(i,0):>7}  peso={w:.4f}')

# ── 8. Dataset e DataLoaders ─────────────────────────────────
# Dataset já augmented → augmentação leve apenas no treino
train_transform = A.Compose([
    A.Resize(IMG_SIZE, IMG_SIZE),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomBrightnessContrast(brightness_limit=0.1, contrast_limit=0.1, p=0.3),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2()
])

val_transform = A.Compose([
    A.Resize(IMG_SIZE, IMG_SIZE),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2()
])


class RetinaDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df        = df.reset_index(drop=True)
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = cv2.imread(row.image)
        if img is None:
            img = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8)
        else:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if self.transform:
            img = self.transform(image=img)['image']
        return img, int(row.label)


train_dataset = RetinaDataset(train_df, train_transform)
val_dataset   = RetinaDataset(val_df,   val_transform)

train_loader = DataLoader(
    train_dataset, batch_size=BATCH_SIZE, shuffle=True,
    num_workers=NUM_WORKERS, pin_memory=False,
    persistent_workers=False,
    prefetch_factor=2 if NUM_WORKERS > 0 else None
)
val_loader = DataLoader(
    val_dataset, batch_size=BATCH_SIZE, shuffle=False,
    num_workers=NUM_WORKERS, pin_memory=False,
    persistent_workers=False,
    prefetch_factor=2 if NUM_WORKERS > 0 else None
)

print(f'\nBatches treino   : {len(train_loader)}')
print(f'Batches validação: {len(val_loader)}')

# ── 9. Modelo ResNet-50 ───────────────────────────────────────
def build_resnet50(n_classes=5, dropout=0.4):
    model       = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
    in_features = model.fc.in_features   # 2048
    model.fc    = nn.Sequential(
        nn.Dropout(p=dropout),
        nn.Linear(in_features, 512),
        nn.ReLU(inplace=True),
        nn.Dropout(p=0.3),
        nn.Linear(512, n_classes)
    )
    return model


def freeze_backbone(model):
    for name, param in model.named_parameters():
        if not name.startswith('fc.'):
            param.requires_grad = False


def unfreeze_all(model):
    for param in model.parameters():
        param.requires_grad = True


model = build_resnet50(n_classes=N_CLASSES, dropout=0.4).to(device)
freeze_backbone(model)

n_total     = sum(p.numel() for p in model.parameters()) / 1e6
n_trainable = sum(p.numel() for p in model.parameters() if p.requires_grad) / 1e6
print(f'\nResNet-50 carregada')
print(f'  Total       : {n_total:.1f} M parâmetros')
print(f'  Treináveis  : {n_trainable:.1f} M (backbone frozen)')

# ── 10. Loss / Optimizer / Scheduler / Scaler ────────────────
criterion = nn.CrossEntropyLoss(weight=class_weights_tensor, label_smoothing=0.05)

optimizer = optim.AdamW(
    filter(lambda p: p.requires_grad, model.parameters()),
    lr=LR, weight_decay=1e-4
)

scheduler = optim.lr_scheduler.CosineAnnealingLR(
    optimizer, T_max=EPOCHS, eta_min=1e-6
)

scaler = GradScaler(enabled=USE_AMP)

print('\n✅ Loss (CrossEntropy+label_smoothing), AdamW, CosineAnnealing, GradScaler prontos')

# ── 11. Checkpoint ───────────────────────────────────────────
start_epoch        = 0
start_batch        = 0
best_val_loss      = float('inf')
early_stop_counter = 0
history            = []


def get_state_dict():
    sd = model.state_dict()
    return {k.replace('_orig_mod.', ''): v for k, v in sd.items()}


def save_checkpoint(epoch, batch, tag=''):
    torch.save({
        'epoch':              epoch,
        'batch':              batch,
        'model':              get_state_dict(),
        'optimizer':          optimizer.state_dict(),
        'scheduler':          scheduler.state_dict(),
        'scaler':             scaler.state_dict(),
        'best_val_loss':      best_val_loss,
        'early_stop_counter': early_stop_counter,
        'history':            history,
    }, CHECKPOINT_PATH)
    if tag:
        print(f'   💾 Checkpoint salvo {tag}')


if os.path.exists(CHECKPOINT_PATH):
    print('\n📂 Checkpoint encontrado — retomando...')
    ckpt = torch.load(CHECKPOINT_PATH, map_location=device)

    sd = {k.replace('_orig_mod.', ''): v for k, v in ckpt['model'].items()}

    ckpt_epoch = ckpt.get('epoch', 0)
    if ckpt_epoch >= FREEZE_EPOCHS:
        unfreeze_all(model)
        optimizer = optim.AdamW(model.parameters(), lr=UNFREEZE_LR, weight_decay=1e-4)
        scheduler = optim.lr_scheduler.CosineAnnealingLR(
            optimizer, T_max=EPOCHS - ckpt_epoch, eta_min=1e-6
        )
        print(f'   ℹ️  Backbone descongelado (epoch {ckpt_epoch} >= {FREEZE_EPOCHS})')

    model.load_state_dict(sd)
    optimizer.load_state_dict(ckpt['optimizer'])
    try:    scheduler.load_state_dict(ckpt['scheduler'])
    except: pass
    if USE_AMP:
        try:    scaler.load_state_dict(ckpt['scaler'])
        except: pass

    start_epoch        = ckpt.get('epoch', 0)
    start_batch        = ckpt.get('batch', 0)
    best_val_loss      = ckpt.get('best_val_loss', float('inf'))
    early_stop_counter = ckpt.get('early_stop_counter', 0)
    history            = ckpt.get('history', [])

    if start_batch >= len(train_loader):
        start_epoch += 1
        start_batch  = 0
    print(f'   Retomando epoch {start_epoch+1} | best_val_loss={best_val_loss:.4f}')
else:
    print('\n🆕 Nenhum checkpoint — treinando do zero')

# ── 12. Funções de validação ─────────────────────────────────
@torch.no_grad()
def validate():
    model.eval()
    total_loss, correct, total = 0.0, 0, 0
    for images, labels in tqdm(val_loader, desc='  Validação', leave=False):
        images = images.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        with autocast(enabled=USE_AMP):
            preds = model(images)
            loss  = criterion(preds, labels)
        total_loss += loss.item()
        correct    += (preds.argmax(1) == labels).sum().item()
        total      += labels.size(0)
        del images, labels, preds, loss
    gc.collect()
    if device.type == 'cuda':
        torch.cuda.empty_cache()
    return total_loss / len(val_loader), correct / total


@torch.no_grad()
def full_metrics():
    model.eval()
    all_labels, all_preds, all_probs = [], [], []
    for images, labels in tqdm(val_loader, desc='  Métricas', leave=False):
        images = images.to(device, non_blocking=True)
        with autocast(enabled=USE_AMP):
            logits = model(images)
        probs = torch.softmax(logits.float(), dim=1).cpu().numpy()
        all_labels.extend(labels.numpy())
        all_preds.extend(probs.argmax(axis=1))
        all_probs.append(probs)
        del images, logits, probs
    gc.collect()
    if device.type == 'cuda':
        torch.cuda.empty_cache()
    return np.array(all_labels), np.array(all_preds), np.vstack(all_probs)

# ── 13. Loop de treinamento ───────────────────────────────────
print(f'\n🚀 Treinando ResNet-50 — epochs {start_epoch+1} → {EPOCHS}')
print(f'   Epochs 1–{FREEZE_EPOCHS}: backbone congelado')
print(f'   Epochs {FREEZE_EPOCHS+1}+: fine-tuning completo')
print('=' * 60)

for epoch in range(start_epoch, EPOCHS):

    # Descongelar backbone na epoch certa
    if epoch == FREEZE_EPOCHS:
        backbone_frozen = not any(
            p.requires_grad
            for n, p in model.named_parameters()
            if not n.startswith('fc.')
        )
        if backbone_frozen:
            unfreeze_all(model)
            optimizer = optim.AdamW(
                model.parameters(), lr=UNFREEZE_LR, weight_decay=1e-4
            )
            scheduler = optim.lr_scheduler.CosineAnnealingLR(
                optimizer, T_max=EPOCHS - epoch, eta_min=1e-6
            )
            n_trainable = sum(p.numel() for p in model.parameters() if p.requires_grad) / 1e6
            print(f'\n🔓 Backbone descongelado! LR={UNFREEZE_LR:.1e} | {n_trainable:.1f} M treináveis')

    model.train()
    total_loss  = 0.0
    epoch_start = time.time()
    status      = '🔒 frozen' if epoch < FREEZE_EPOCHS else '🔓 full'

    pbar = tqdm(
        enumerate(train_loader), total=len(train_loader),
        desc=f'Epoch {epoch+1}/{EPOCHS} [{status}]', dynamic_ncols=True
    )

    for batch_idx, (images, labels) in pbar:

        if epoch == start_epoch and batch_idx < start_batch:
            continue

        images = images.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        with autocast(enabled=USE_AMP):
            preds = model(images)
            loss  = criterion(preds, labels)

        scaler.scale(loss).backward()

        # Gradient clipping
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()
        avg_loss    = total_loss / (batch_idx - start_batch + 1)
        del images, labels, preds, loss

        pbar.set_postfix({
            'loss': f'{avg_loss:.4f}',
            'lr':   f'{optimizer.param_groups[0]["lr"]:.2e}'
        })

        if (batch_idx + 1) % SAVE_EVERY_N_BATCHES == 0:
            save_checkpoint(epoch, batch_idx + 1,
                            tag=f'(epoch {epoch+1}, batch {batch_idx+1})')

    start_batch = 0

    val_loss, val_acc = validate()
    scheduler.step()

    elapsed = time.time() - epoch_start
    print(f'\nEpoch {epoch+1}/{EPOCHS} [{status}] | '
          f'Train Loss: {avg_loss:.4f} | '
          f'Val Loss: {val_loss:.4f} | '
          f'Val Acc: {val_acc:.4f} | '
          f'{elapsed/60:.1f} min')

    history.append({
        'epoch':      epoch + 1,
        'train_loss': avg_loss,
        'val_loss':   val_loss,
        'val_acc':    val_acc,
        'frozen':     epoch < FREEZE_EPOCHS
    })

    if val_loss < best_val_loss:
        best_val_loss      = val_loss
        early_stop_counter = 0
        torch.save(get_state_dict(), FINAL_MODEL)
        print(f'   ⭐ Melhor modelo salvo! Val Loss={val_loss:.4f} | Acc={val_acc:.4f}')
    else:
        early_stop_counter += 1
        print(f'   ⚠️  Sem melhoria ({early_stop_counter}/{EARLY_STOP_PATIENCE})')
        if early_stop_counter >= EARLY_STOP_PATIENCE:
            print('🛑 Early Stopping!')
            save_checkpoint(epoch, len(train_loader), tag='(early stop)')
            break

    save_checkpoint(epoch + 1, 0, tag=f'(fim epoch {epoch+1})')

    gc.collect()
    if device.type == 'cuda':
        torch.cuda.empty_cache()

    print('=' * 60)

print('\n🏁 Treino ResNet-50 finalizado!')

# ── 14. Gráficos de treino ───────────────────────────────────
if history:
    df_h         = pd.DataFrame(history)
    epochs_done  = df_h['epoch'].tolist()
    train_losses = df_h['train_loss'].tolist()
    val_losses   = df_h['val_loss'].tolist()
    val_accs     = df_h['val_acc'].tolist()
    freeze_end   = (df_h[~df_h['frozen']]['epoch'].min()
                    if (~df_h['frozen']).any() else FREEZE_EPOCHS)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    ax1.plot(epochs_done, train_losses, 'o-', label='Train Loss', color='#2E75B6')
    ax1.plot(epochs_done, val_losses,   'o-', label='Val Loss',   color='#DD8452')
    if freeze_end and freeze_end < max(epochs_done):
        ax1.axvline(x=freeze_end - 0.5, color='gray', ls='--', alpha=0.7,
                    label=f'Unfreeze (epoch {freeze_end})')
    ax1.set_title('Loss por Epoch — ResNet-50', fontweight='bold')
    ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss')
    ax1.legend(); ax1.grid(alpha=0.3)

    ax2.plot(epochs_done, val_accs, 'o-', label='Val Accuracy', color='#2CA02C')
    if freeze_end and freeze_end < max(epochs_done):
        ax2.axvline(x=freeze_end - 0.5, color='gray', ls='--', alpha=0.7,
                    label=f'Unfreeze (epoch {freeze_end})')
    ax2.set_title('Acurácia de Validação — ResNet-50', fontweight='bold')
    ax2.set_xlabel('Epoch'); ax2.set_ylabel('Accuracy')
    ax2.legend(); ax2.grid(alpha=0.3)

    plt.suptitle('ResNet-50 | Retinopatia Diabética | augmented_resized_V2',
                 fontsize=13, fontweight='bold')
    plt.tight_layout()
    path_curves = os.path.join(OUTPUT_DIR, 'resnet50_curvas_treino.png')
    plt.savefig(path_curves, dpi=150, bbox_inches='tight')
    plt.show()
    print(f'💾 Salvo: {path_curves}')

    print('\nHistórico:')
    print(df_h[['epoch','train_loss','val_loss','val_acc']].to_string(index=False))

# ── 15. Métricas finais ──────────────────────────────────────
print('\n📂 Carregando melhor modelo para avaliação final...')
sd = torch.load(FINAL_MODEL, map_location=device)
sd = {k.replace('_orig_mod.', ''): v for k, v in sd.items()}
model.load_state_dict(sd)
print('✅ Melhor modelo carregado')

labels_arr, preds_arr, probs_arr = full_metrics()

acc   = accuracy_score(labels_arr, preds_arr)
kappa = cohen_kappa_score(labels_arr, preds_arr, weights='quadratic')
labels_bin = label_binarize(labels_arr, classes=list(range(N_CLASSES)))
try:
    auc = roc_auc_score(labels_bin, probs_arr, average='macro', multi_class='ovr')
except:
    auc = float('nan')

print(f'\n{"="*55}')
print('📊 MÉTRICAS FINAIS — ResNet-50')
print(f'{"="*55}')
print(f'  Acurácia        : {acc:.4f} ({acc*100:.1f}%)')
print(f'  Kappa Quadrático: {kappa:.4f}')
print(f'  AUC-ROC (macro) : {auc:.4f}')
print(f'{"="*55}')
print('\n📋 Classification Report:')
print(classification_report(labels_arr, preds_arr,
                             target_names=CLASS_NAMES, digits=4))

# ── 16. Matriz de Confusão ───────────────────────────────────
cm      = confusion_matrix(labels_arr, preds_arr)
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True).clip(min=1)

fig, ax = plt.subplots(figsize=(8, 6.5))
sns.heatmap(cm_norm, annot=False, ax=ax, cmap='Blues', vmin=0, vmax=1,
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES,
            linewidths=0.5, linecolor='white')

for i in range(N_CLASSES):
    for j in range(N_CLASSES):
        val   = cm_norm[i, j]
        color = 'white' if val > 0.5 else 'black'
        ax.text(j+0.5, i+0.38, f'{val:.2f}', ha='center', va='center',
                fontsize=10, fontweight='bold', color=color)
        ax.text(j+0.5, i+0.65, f'n={cm[i,j]}', ha='center', va='center',
                fontsize=7.5, color=color, alpha=0.8)

ax.set_title(f'Matriz de Confusão — ResNet-50\n'
             f'Acc={acc:.4f}  |  Kappa={kappa:.4f}  |  AUC={auc:.4f}',
             fontweight='bold', fontsize=12)
ax.set_ylabel('Classe Real', fontsize=11)
ax.set_xlabel('Classe Predita', fontsize=11)
ax.tick_params(axis='x', rotation=30)
ax.tick_params(axis='y', rotation=0)

plt.tight_layout()
path_cm = os.path.join(OUTPUT_DIR, 'resnet50_confusion_matrix.png')
plt.savefig(path_cm, dpi=150, bbox_inches='tight')
plt.show()
print(f'💾 Salvo: {path_cm}')

# ── 17. Verificar arquivos ───────────────────────────────────
print('\n📁 Arquivos gerados:')
for path, label in [
    (FINAL_MODEL,    'Modelo final'),
    (CHECKPOINT_PATH,'Checkpoint'),
]:
    if os.path.exists(path):
        print(f'  ✅ {label}: {path} ({os.path.getsize(path)/1e6:.0f} MB)')
    else:
        print(f'  ❌ {label}: não encontrado')

for fname in ['resnet50_curvas_treino.png', 'resnet50_confusion_matrix.png']:
    full = os.path.join(OUTPUT_DIR, fname)
    print(f'  {"✅" if os.path.exists(full) else "❌"} {fname}')

print(f'\n📂 Resultados em: {OUTPUT_DIR}')
```
